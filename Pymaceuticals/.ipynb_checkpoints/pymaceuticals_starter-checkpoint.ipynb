## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.

myMean = myDf.groupby("Drug Regimen").mean()["Tumor Volume (mm3)"]
myMedian = myDf.groupby("Drug Regimen").median()["Tumor Volume (mm3)"]
myVariance = myDf.groupby("Drug Regimen").var()["Tumor Volume (mm3)"]
myStdDev = myDf.groupby("Drug Regimen").std()["Tumor Volume (mm3)"]
myStdError = myDf.groupby("Drug Regimen").sem()["Tumor Volume (mm3)"]

# Create Dataframe
ssTable = pd.DataFrame({
    "Mean Tumor Volume": myMean,
    "Median Tumor Volume": myMedian,
    "Tumar Volume Variance": myVariance,
    "Tumor Volume Std. Dev.": myStdDev,
    "Tumor Volume Std. Error": myStdError
})

#Preview the dataframe
ssTable
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line
singleSSTable = myDf.groupby("Drug Regimen").agg({"Tumor Volume (mm3)": ["mean", "median", "var", "std", "sem"]})
singleSSTable
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# set drug regimen as index and drop associated regimens while only keeping Capomulin, Ramicane, Infubinol, and Ceftamin
myDrugDf = myDf.set_index('Drug Regimen')
myDrugDf = myDrugDf.loc[['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin'],:]

# isolated view of just capomulin for later use
capomulin  = myDrugDf.loc['Capomulin',:]

# Reset index so drug regimen column persists after inner merge
myNewDrugDf = myDrugDf.reset_index()

# get mouse count per drug
mouseDrugCount = myNewDrugDf['Drug Regimen'].value_counts()
print(mouseDrugCount)

# Start by getting the last (greatest) timepoint for each mouse
myTimepoint = myNewDrugDf.groupby('Mouse ID')['Timepoint'].max()
myTimepoint.head()
```

```python
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
newDf = myTimepoint.reset_index().merge(myDf, on=['Mouse ID', 'Timepoint'], how='left')
newDf = newDf.set_index('Mouse ID')

# show all rows of data
newDf = newDf.sort_values(by='Timepoint')
newDf = newDf[['Drug Regimen', 'Timepoint', 'Tumor Volume (mm3)', 'Metastatic Sites', 'Sex', 'Age_months', 'Weight (g)']]
pd.set_option('display.max_rows', None)
newDf
```

```python
# Put treatments into a list for for loop (and later for plot labels)
print(newDf['Drug Regimen'].unique())
print("-------------")
treatmentsList = ['Ceftamin', 'Infubinol', 'Ramicane', 'Capomulin']

# Create empty list to fill with tumor vol data (for plotting)
tumorVol = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for x in treatmentsList:
    
    tumorCount = newDf.loc[newDf['Drug Regimen'] == x, 'Tumor Volume (mm3)']
    tumorVol.append(tumorCount)
    
    qtls = tumorCount.quantile([0.25, 0.5, 0.75])
    lowerQ = qtls[0.25]
    upperQ = qtls[0.75]
    IQR = upperQ - lowerQ
    lowerBound = lowerQ - (1.5 * IQR)
    upperBound = upperQ + (1.5 * IQR)
    outlier = tumorCount.loc[(tumorCount < lowerBound) | (tumorCount > upperBound)]
    print(outlier)
print("-------")

# Locate the rows which contain mice on each drug and get the tumor volumes

print(tumorVol)
```

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
#establish x and y values and find St. Pearson Correlation Coefficient for Mouse Weight and Tumor Volume Avg
weight = capomulinAvg['Weight (g)']
tumorVolumemm3 = capomulinAvg['Tumor Volume (mm3)']
corrlation = st.pearsonr(weight,tumorVolumemm3)

#print St. Pearson Correlation Coefficient
print(f'The correlation between both factors is {round(corrlation[0],2)}')

# establish linear regression values
(slope, intercept, rvalue, pvalue, stderr) = linregress(weight, tumorVolumemm3)
regressionValues = weight * slope + intercept


# linear regression line 
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

# scatter plot of the data
plt.scatter(weight, tumorVolumemm3)
plt.plot(weight, regressionValues, "r-")
plt.xlabel('Mouse Weight (g)')
plt.ylabel('Tumor Volume Average')
plt.title('Tumor Volume Average vs. Mouse Weight Linear Regression')
plt.show()
#This hw was not fun :/
```
